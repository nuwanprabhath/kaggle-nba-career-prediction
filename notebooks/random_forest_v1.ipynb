```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
train_df = pd.read_csv('../data/raw/train.csv')
test_df = pd.read_csv('../data/raw/test.csv')
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
train_set = train_df.loc[:, 'GP':'TOV'] # Selecting all columns except Id and TARGET_5Yrs as training set
target_set = train_df.loc[:, 'TARGET_5Yrs'] # Selecting TARGET_5Yrs as the target
test_set = test_df.loc[:, 'GP':'TOV']
```

```python
clf = RandomForestClassifier(n_estimators=10, n_jobs=1)
```

```python
clf.fit(train_set, target_set)
```

```python
pred = clf.predict(test_set)
id_df = test_df.loc[:,'Id']

result  = {'Id': id_df, 'TARGET_5Yrs': pred} 
result = pd.DataFrame(data = result)
result.to_csv("submission_v1.csv", index = False)
```
