## Environment Setup

```python
import pandas as pd
from importlib.machinery import SourceFileLoader
dataprep = SourceFileLoader('prep', '../src/data/prepare.py').load_module()
sets = SourceFileLoader('sets', '../src/data/sets.py').load_module()
from sklearn.preprocessing import MinMaxScaler
```

# 2. Data Preparation

**<u>Collect Initial Data</u>**

```python
df_train = pd.read_csv('../data/raw/train.csv')
target_name = 'TARGET_5Yrs'
```

---
**<u>Create new dataframe for data cleansing</u>**

```python
df_cleaned = df_train.copy()
```

---
**<u>Construct New Features</u>**

---
**<u>Drop Unnessary Features</u>**

```python
df_cleaned = dataprep.drop_features(df_cleaned, ['Id'])
df_cleaned.info()
```

---
**<u>Scale Features</u>**

```python
df_cleaned = dataprep.scale_features(df_cleaned, MinMaxScaler(), target_name)
```

```python
df_cleaned.describe()
```

---
**<u>Split dataset into Train and Validation sets</u>**

```python
X_train, y_train, X_val, y_val = dataprep.split_sets(df_cleaned, target_name)
```

```python
X_train.describe()
```

```python
X_val.describe()
```

---
**<u>Print observation count and Frequency of target variables for split datasets</u>**

```python
print('Observations in train dataset = ' + str(y_train.shape))
print('Frequency of target variable in train dataset')
y_train.value_counts(normalize=True)
```

```python
print('Observations in validation dataset = ' + str(y_val.shape))
print('Frequency of target variable in validation dataset')
y_val.value_counts(normalize=True)
```

**Observations:**
* Distribution of target variable consistent across train and validation datasets and consistent with original dataset

---
**<u>Save datasets for modelling</u>**

```python
sets.save_sets(X_train, y_train, X_val, y_val)
```
